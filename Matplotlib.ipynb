```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
plt.plot(x_values,y_values)
plt.text(1,1,"SMIT CLASS")
plt.title("SMIT CLASS")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.show()
```

```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
y2_values=[2,3,6,12,18,27]
plt.plot(x_values,y_values,label="SMIT",c="red",marker="*",linestyle="--")
plt.plot(x_values,y2_values,label="CLASS",c="Blue")
plt.text(1,1,"SMIT CLASS")
plt.title("SMIT CLASS")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.ylim(0,50)
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
x_values=[0,1,2,3,4,5]
y_values=[0,1,4,9,16,25]
y2_values=[2,3,6,12,18,27]
plt.plot(x_values,y_values,label="SMIT",c="Green",marker="*",linestyle="--")
plt.plot(x_values,y2_values,label="CLASS",c="Blue")
plt.text(1,1,"SMIT CLASS")
plt.title("SMIT CLASS")
plt.xlabel("x_label")
plt.ylabel("y_label")
plt.ylim(-50,50)
plt.xlim(-10,10)
plt.axhline(0,c="red")
plt.axvline(0,c="Black")
plt.legend()
plt.show()
```
