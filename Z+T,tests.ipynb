```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as smt
import seaborn as sns
```

```python
#pip install scipy
```

```python
x_bar=980
n=50
s_mean=1000
p_s=50
z_test=(x_bar-s_mean)/(p_s/np.sqrt(n))
print(z_test)

x=np.linspace(-4,4,500)
y=smt.norm.pdf(x,0,1)
z_crit=smt.norm.ppf(1-0.05/2)
plt.plot(x,y)
plt.fill_between(x,y,where=(x<=-z_crit),color="red",label="Rejection H0")
plt.fill_between(x,y,where=(x>=z_crit),color="red",label="Rejection H0")
plt.axvline(z_test,color="g",label="Z-Test Score lie")
print(z_crit)
print(z_test)
plt.legend()
```

```python
n=10
x_bar=12
p_mean=12.1
s=1.135
df=n-1
t_test=(p_mean-x_bar)/(s/np.sqrt(n))
t_crit=smt.t.ppf(1-0.05/2,df)
x=np.linspace(-4,4,500)
y=smt.t.pdf(x,df)
plt.plot(x,y)
plt.fill_between(x,y,where=(x<=-t_crit),color="red",label="Rejection H0")
plt.fill_between(x,y,where=(x>=t_crit),color="red",label="Rejection H0")
plt.axvline(t_test,color="g",label="t-Test Score lie")
print(t_crit)
print(t_test)
plt.legend()
```
