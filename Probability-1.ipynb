```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ss=["red"]*3+["blue"]*5+["green"]*2
```

```python
ss=pd.Series(ss)
```

```python
print(f"Theoritical Probability of Blue Ball is:{ss.value_counts()["blue"]/len(ss)}")
print(f"Theoritical Probability of red Ball is:{ss.value_counts()["red"]/len(ss)}")
print(f"Theoritical Probability of green Ball is:{ss.value_counts()["green"]/len(ss)}")
```

```python
ss1=np.random.choice(["red","blue","green"],size=10000)
ss1=pd.Series(ss1)
```

```python
print(f"Emperical Probability of blue is:{ss1.value_counts(normalize=True)["blue"]:.2f}")
print(f"Emperical Probability of  red is:{ss1.value_counts(normalize=True)["red"]:.2f}")
print(f"Emperical Probability of  green is:{ss1.value_counts(normalize=True)["green"]:.2f}")
```

```python
for ball in ss.unique():
    print(f"Emperical Probability of {ball} Ball is:{ss1.value_counts(normalize=True)[ball]:.2f}")
```
