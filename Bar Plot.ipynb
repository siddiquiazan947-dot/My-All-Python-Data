```python
Student_Data=pd.read_csv(r"C:\\Users\\H S COMPUTER S HYD\\Desktop\\student_data.csv")
```

```python
Student_Data["Obtain_Marks"]=Student_Data.iloc[:,2:5].sum(axis=1)
```

```python
Student_Data["Total_Marks"]=100
```

```python
Student_Data["Percentage"]=(Student_Data.Obtain_Marks/Student_Data.Total_Marks)*100
```

```python
def grade(marks):
    if marks>80:
        return"A"
    elif marks>70:
        return"B"
    elif marks>50:
        return"C"
    else:
        return"F"
Student_Data["Grade"]=Student_Data.Percentage.apply(grade)
```

```python
Class=Student_Data["roll_no"].str.split("/").str[1]
Student_Data["Class"]=Class
Student_Data
```

```python
cl_group=Student_Data.Class.unique()
cl_group
```

```python
for cl in cl_group:
    group=Student_Data.groupby("Class")
    call_group=group.get_group(cl)
    x=call_group.Grade.value_counts().index
    y=call_group.Grade.value_counts().values
    print(cl)
    plt.title(cl)
    plt.bar(x,y,color=("r","y","b","g"),alpha=0.9)
    plt.show()
```
